```python
from datasets import concatenate_datasets, load_dataset
import os
import random
from tqdm import tqdm
```

```python
bookcorpus = load_dataset("bookcorpus", split="train")
```

```python
bookcorpus[0]
```

```python
wiki = load_dataset("wikipedia", "20220301.en", split="train")
wiki = wiki.remove_columns([col for col in wiki.column_names if col != "text"])  # only keep the 'text' column
```

```python
wiki[0]
```

```python
dataset = concatenate_datasets([bookcorpus, wiki])
```

```python
dataset[0]
```

```python
seed = 264806
random.seed(seed)
```

```python
max_file_size = 2.52 * 1024 * 1024 * 1024
filename = f"FINAL_CORPUS_SEED_{seed}.txt"
file_size = 0
pb = tqdm(total=max_file_size)

written = set()
while file_size <=max_file_size:
    indices = random.sample(range(len(dataset)), 20000)
    strings = dataset.select(indices)
    split_strings = [s['text'].split('\n') for s in strings]
    to_write = [x for xs in split_strings for x in xs]
    with open(filename, 'a', encoding='utf-8') as file:
        for s in to_write:
            cleaned = s.strip()
            if len(cleaned) < 1 or cleaned.isspace(): continue
            if cleaned not in written:
              file.write(cleaned + '\n')
              written.add(cleaned)

    file_size =os.path.getsize(filename)
    pb.update(file_size)
```

```python
lines = open(filename, 'r').read().splitlines()
print(f"Final file size = {os.path.getsize(filename) / 1024 / 1024 / 1024}GB")
print(f"The reduced corpus contains {len(lines)} pieces of text")
```
